```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
AzerUrl = 'https://en.wikipedia.org/wiki/2019_Azerbaijan_Grand_Prix'
```

```python
responseAzer = requests.get(AzerUrl)
```

```python
tablesAzer = pd.read_html(AzerUrl)
```

```python
f1AzerTableFinale = tablesAzer[3]
```

```python
f1AzerTableFinale
```

```python
AzerTopTen = f1AzerTableFinale.head(10)
```

```python
AzerTop10 = AzerTopTen.drop("No.", axis=1)
```

```python
AzerTop10new = AzerTop10.drop("Points", axis=1)
```

```python
AzerTop10new
```

```python
Top10Azer = AzerTop10new.drop("Laps", axis=1)
```

```python
Top10Azer
```

```python
Top10Azer.to_html('/Users/christinemagnotta/FormulaOneGroup/top10Azer.html')
```
